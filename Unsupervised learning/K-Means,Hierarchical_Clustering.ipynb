## Preprocessing

```python
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

## K-means 

```python
from sklearn.cluster import KMeans

# Define range of cluster numbers
cluster_range = range(2, 11)  

inertia = []

# Perform K-means clustering for each number of clusters
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    
    # Append inertia
    inertia.append(kmeans.inertia_)

# Plot Inertia
plt.figure(figsize=(12, 5))
plt.plot(cluster_range, inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Inertia vs Number of Clusters')
plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
```

```python
cluster=kmeans.labels_

df["cluster_no"]=cluster
df["cluster_no"].value_counts()
```

```python
from sklearn.decomposition import PCA


kmeans = KMeans(n_clusters=3, random_state=0)

labels = kmeans.fit_predict(X_scaled)

# Perform PCA to reduce the data to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Get the cluster centers in the PCA-reduced space
centers_pca = pca.transform(kmeans.cluster_centers_)

plt.figure(figsize=(10, 7))

# Plot the data points with colors based on cluster labels
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels,  label='Data Points')

# Plot the cluster centers
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red',  marker='X', label='Cluster Centers')

plt.title('K-Means Clustering (PCA-reduced Data)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()

plt.show()
```

## Hierarchical clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

linked = linkage(X_scaled, method='ward')  # 'ward' minimizes variance within clusters

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```
